```python
import torch 
import torch.nn as nn
import torchvision
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import torch.optim as optim
```

```python
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()
```

```python
from torch.utils.data import Dataset, DataLoader
```

```python
x_train=x_train.reshape(-1,28*28)
```

```python
x_test=x_test.reshape(-1,28*28)
```

```python
data=[]
val=[]
for i in range(len(x_train)):
    val=[]
    val.append(x_train[i])
    val.append(y_train[i])
    data.append(val)
    
```

```python
data2=[]

for i in range(len(x_test)):
    val=[]
    val.append(x_test[i])
    val.append(y_test[i])
    data2.append(val)
```

```python
X_train=DataLoader(data,batch_size=100,shuffle=True)
```

```python
X_test=DataLoader(data2,batch_size=100,shuffle=True)
```

```python
class net(nn.Module):
    def __init__(self):
        super().__init__()
        self.fc1=nn.Linear(28*28,400)
        self.fc2=nn.Linear(400,800)
        self.fc3=nn.Linear(800,1600)
        self.fc4=nn.Linear(1600,10)
        self.rel=nn.ReLU()
    def forward(self,x):
        x=self.rel(self.fc1(x))
        x=self.rel(self.fc2(x))
        x=self.rel(self.fc3(x))
        x=self.fc4(x)
        return x
        
```

```python
neur=net()
```

```python
criterion=nn.CrossEntropyLoss()
optimizer=optim.Adam(neur.parameters(),lr=0.001)
```

```python
epoch=10
```

```python
for i in range(epoch):
    for data,label in X_train:
        data=torch.tensor(data,dtype=torch.float32)
        out=neur(data)
        loss=criterion(out,label)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        print(f"Epoch {i}, Loss: {loss.item():.6f}")
```

```python
neur.eval()  # set model to evaluation mode (no dropout/batchnorm updates)
correct=0
total=0
with torch.no_grad():  # disable gradient computation for faster inference
    for data, label in X_test:
        data = data.float()
        label = label.long()  # CE loss expects long (int64) class labels

        # Forward pass
        outputs = neur(data)  # shape: [batch, 10]

        # Get predicted class (highest logit)
        _, predicted = torch.max(outputs, 1)  # returns (max_value, index)

        # Update counts
        total += label.size(0)
        correct += (predicted == label).sum().item()
```

```python
print((correct/total)*100)
```
